### Efficiency Maps Scan

```python
tau_vals = np.logspace(-2.5,1,10).tolist()
mGo_vals = np.arange(500.,1500.,250.).tolist() + np.arange(1500.,3000.,200.).tolist()
mN1_vals = np.arange(50.,2900.,200.).tolist()
print(len(tau_vals),len(mGo_vals),len(mN1_vals))
```

```python
scanPts = []
for mn1,mgo,tau in itertools.product(mN1_vals,mGo_vals,tau_vals):
    if mn1 >= mgo or mn1 <= 0:
        continue
    scanPts.append((mn1,mgo,tau))
scanPts = np.array(scanPts)
print(len(scanPts))
ptsIni = [("%1.2f" %pt[0], "%1.2f" %pt[1], "%1.2e" %pt[2]) for pt in scanPts]
# ptsIni = sorted([("%1.2f" %pt[0], "%1.2f" %pt[1], "%1.2e" %pt[2], "75000") for pt in ptsSel])
print(str(ptsIni).replace("'", ""))
```

```python
plt.figure(figsize=(8,5))
plt.scatter(scanPts[:,1],scanPts[:,2],label='Points')
# plt.legend()
# plt.xscale('log')
# plt.xlim(0,1150)
# plt.ylim(4.,6.)
plt.xlabel(r'$m_{\tilde g}$ (GeV)')
plt.ylabel(r'$\tau$ (ns)')
plt.yscale('log')
# plt.title(r'$N = %i$' %len(ptsNew))
plt.show()
```
