```python
import numpy as np
from matplotlib import pyplot as plt
from itertools import product
```

## Strong Production

### Fig.12

```python
atlasCurve = np.genfromtxt('../ATLAS_data/HEPData-ins2628398-v1-csv/excl_strong_mgluino_2400_GeV_obs_nominal.csv',delimiter=',',comments='#')
```

```python
mN1 = np.append(np.arange(50.,200.,25.),np.arange(200.,1900.,200.))
tau = np.logspace(-3.,1.0,20)
mGluino = [2400.]
pts = np.array(list(product(mN1,mGluino,tau)))
plt.scatter(pts[:,2],pts[:,0])
plt.plot(atlasCurve[:,0],atlasCurve[:,1],linewidth=3,c='gray',linestyle='dashed')
plt.xscale('log')
plt.xlim(1e-3,1e2)
plt.ylim(25.,2000.)
plt.xlabel(r'$\tau$ (ns)')
plt.ylabel(r'$m_{\tilde \chi_1^0}$ (GeV)')
plt.title(r'$N = %i$' %len(pts))
plt.show()
```

```python
ptsIni = [("%1.0f" %pt[0],"%1.0f" %pt[1], "%1.3e" %pt[2]) for pt in pts]
```

```python
print(str(ptsIni).replace("'", ""))
```
